# 7. correlation(linear) b/w target and features

```python
x = train.corr()['compliance_2021'].sort_values(ascending= False)
x
```

```python
x[x.values>0.5]
```

with threshold 50%(0.5) there are no feature linearly correalted with target variable.

```python
# correlation between features
def corrFilter(bound: float):
    xCorr = train.corr()#["y_yes"]
    data_corr_Filtered = xCorr[((xCorr >= bound) | (xCorr <= -bound)) & (xCorr !=1.000)]
    data_corr_Flattened = data_corr_Filtered.unstack().sort_values().drop_duplicates()
    return data_corr_Flattened
corrFilter(0.7)
```

with threshold 70%, it seems all the features of 2019 has high correaltion with features of 2020
